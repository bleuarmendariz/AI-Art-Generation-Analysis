## Step 2: Extracting data from string patterns

Next we extract the data from inbetween the ** and **. Remember this contains our users query - what they are actually trying to create using the AI.

```python
def clean_astr(x):
    start = '**'
    
    #Using string slicing to slice for data in between ** and **
    clean_string = x[x.find(start) + 2:]
    clean_string = clean_string[:clean_string.find('*')]
    clean_string = clean_string.lower().strip().replace(",","")
    return clean_string

#Apply our clean string method to our Content column
df['Content'] = df['Content'].apply(clean_astr)
df.reset_index()
```

# Some Cleaning and Prepping

Now that we have the cleaning done we can get back to answering our question: 

**What words are being used to generate AI art?**

I think we can answer this question with 2 types of measures:
- Counts of most frequently used unique words used (excluding stopwords -more info later)
- What are the most frequently associated words to our popular words?

```python
content = df['Content'].tolist()
```

```python
def word_counter(x):
    word_count = {}

    for i in x:
        for j in i.split():
            if j in word_count:
                word_count[j] += 1
            else:
                word_count[j] = 1
    return word_count
```

```python
word_count = pd.DataFrame.from_dict([word_counter(content)]).transpose().sort_values(0, ascending = False)
```

```python
word_count.head(20)
```

## Discovering our first insight!

In the code above, I iterate through all the strings of user input, and create a simple count of all the words. The immediate problem we see are common words like a, on, the, and in. We can also see a lot of descriptor words that are popular to feed the AI, that will influence the art style. Example: realistic, 4k, 8k, render. 

The descriptors are pretty cool information in itself here - its telling us what kind of styles users want the AI to generate, and it overwhelmingly seems to be: people want realistic images.

Notice how we dont see any counts of -anime style, cartoon, comic book, painting. We see realistic, cinimatic, 4k, and detailed. There is our first insight!

## Dealing with Stopwords and Descriptors

My experience with NLP is very limited, but I dont know about stopwords, and knew this would likely be a problem. Stopwords, if you dont know, are very popular words used in language, that dont contribute a lot of meaning or context. I work with NLTK to remove the stopwords from our content column, and also manually remove some of what I consider "artificial" stop words - the descriptors that are being fed to the AI to specify an art style

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Here we are tokenize our strings in the content column, and then check if those words are in NLTK's stop words dictionary
stop_words = set(stopwords.words('english'))
content_tokenized = []
for i in content:
    word_tokens = word_tokenize(i)
    for w in word_tokens:
        if w not in stop_words:
            content_tokenized.append(w)
```

```python
#We run our strings that have been tokenized and removed of stop words through our simple word counter
content_rem_stops = word_counter(content_tokenized)
```

```python
content_rem_stops = pd.DataFrame.from_dict([word_counter(content_tokenized)]).transpose().sort_values(0, ascending = False)
```

```python
content_rem_stops
```

We see a strange mistake that happened here when tokenizing words... We have a ton of special characters and symbols. I dont see a ton of value in keeping them... Lets get them removed.

```python
content_rem_stops = content_rem_stops.rename(columns = {0:"Count"})
```

```python
content_rem_stops.reset_index(inplace = True)
```

```python
content_rem_stops = content_rem_stops.rename(columns = {'index':'Word'})
```

```python
# The above code just cleans our data frame up to make it easier to work with. Below is the important code. We use a regex pattern to filter for special character strings, and remove them
content_rem_stops.drop(content_rem_stops[content_rem_stops['Word'].str.contains(r'[0-9a-zA-Z]') == False].index, inplace = True)
```

```python
content_rem_stops.head(15)
```

To clarify, what we want to do is take our top 20 list of words, find all of the input from users that contain that word, create a count of all of those words, and plot it like we did above.
```python
most_pop_words = top_20_words['Word'].str.lower().tolist()
most_pop_words
```

Below im going to get the most popular words that are associated with the most popular words... remove the stop words, and concatinate into one dataframe with a label.

```python
# A horribly written way to get the most popular words associated with our most popular words, remove stop words, and concatinate into one dataframe.
most_pop_words_df = pd.DataFrame()
for i, y in enumerate(most_pop_words):
    keyword_df = df[df['Content'].str.contains(most_pop_words[i])]
    print(y)
    keyword_series = keyword_df['Content'].tolist()
    content_tokenized = []
    for j in keyword_series:
        word_tokens = word_tokenize(j)
        for w in word_tokens:
            if w not in stop_words:
                content_tokenized.append(w)
    df_count = pd.DataFrame.from_dict([word_counter(content_tokenized)]).transpose().sort_values(0, ascending = False).iloc[:50]
    df_count['most_pop_word'] = y
    df_count = df_count.rename(columns = {0:'Count'})
    df_count = df_count.reset_index()
    df_count = df_count.rename(columns = {'index':'Word'})
    if i == 0:
        most_pop_words_df = df_count
    else:
        most_pop_words_df = pd.concat([most_pop_words_df, df_count])
    print("Completed {} Dataframe".format(i + 1))
   
```
